```python
import pandas as pd
import ast
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv(".\simulation_data.csv")
df
```

```python
# removing those entries that have (none,none)
df = df[df['Standard Deviation between Vehicle Coordinates'] != "(None, None)"]
df
```

```python
df[['std_dev_x', 'std_dev_y']] = df['Standard Deviation between Vehicle Coordinates'].apply(lambda x: pd.Series(ast.literal_eval(x)))
df[['mean_coord_x', 'mean_coord_y']] = df['Mean Position of Vehicle Coordinates'].apply(lambda x: pd.Series(ast.literal_eval(x)))
```

```python
df
```

```python
# going by the logic that the vehicles will be closer at highly congested places 
# so the standard devaition between the coordinates of vehicles will be less at highly congested places
# we'll be extracting only those timestamps so that its easier for the LLM to analyse 
high_congestion_df = df[(df['std_dev_x'] < df['std_dev_x'].quantile(0.1)) | (df['std_dev_y'] < df['std_dev_y'].quantile(0.1))]
high_congestion_df.drop(columns=['Standard Deviation between Vehicle Coordinates','Mean Position of Vehicle Coordinates', 'Position of Vehicles'], inplace=True)
high_congestion_df.reset_index(drop=True, inplace=True)
high_congestion_df.head()
```

```python
high_congestion_df['Standard Deviation between Vehicle Coordinates'] = list(zip(high_congestion_df['std_dev_x'],high_congestion_df['std_dev_y']))
high_congestion_df['Mean Position of Vehicle Coordinates'] = list(zip(high_congestion_df['mean_coord_x'], high_congestion_df['mean_coord_y']))

high_congestion_df.drop(['std_dev_x', 'std_dev_y', 'mean_coord_x', 'mean_coord_y'], axis=1, inplace=True)
```
